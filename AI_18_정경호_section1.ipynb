```python
def convert_sales(sales):
    if "K" in sales:
        return float(sales.replace("K", "")) * 0.001
    elif "M" in sales:
        return float(sales.replace("M", ""))
    else:
        return float(sales)
        
cdf["NA_Sales"] = cdf["NA_Sales"].apply(convert_sales)
```

```python
cdf["EU_Sales"] = cdf["EU_Sales"].apply(convert_sales)
```

```python
cdf["JP_Sales"] = cdf["JP_Sales"].apply(convert_sales)
```

```python
cdf["Other_Sales"] = cdf["Other_Sales"].apply(convert_sales)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.suptitle('Total Sales by Genre (Top 3) in 5-year period')

colors = ['#ff9999', '#66b3ff', '#99ff99']

min_year = cdf1_year_top3_genre['Year'].min()
max_year = cdf1_year_top3_genre['Year'].max()

num_subplots = (max_year - min_year) // 5 + 1

for i, year in enumerate(sorted(cdf1_year_top3_genre['Year'].unique())):
    df = cdf1_year_top3_genre[cdf1_year_top3_genre['Year'] == year]
    ax = plt.subplot(2, num_subplots, i+1)
    ax.pie(df['Total_Sales'], labels=df['Genre'], colors=colors, autopct='%.1f%%', startangle=90)
    ax.axis('equal')
    ax.set_title(f'{year}s')

plt.tight_layout()
plt.show()
```
